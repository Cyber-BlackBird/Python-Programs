**Write a program to print twin primes  less than 1000. If two consecutive odd numbers are both prime then they are known as twin primes**

```
def prime(num):
  if num>1:
    for i in range(2,num):
      if num%i==0:            
        break
    else:        
      return True



for n in range(1001):
  if prime(n) == True:
    m = n + 2
    if prime(m) == True:
      print('{} & {} are twin primes'.format(n,m))
    
```

**Write a program to implement these formulae of permutations and combinations. Number of permutations of n objects taken r at a time: p(n, r) = n! / (n-r)!.  Number of combinations of n objects taken r at a time is: c(n, r) = n! / (r!*(n-r)!) = p(n,r) / r!**

```
#Factorial of a number
def fact(a):
  if a == 1:
    return a
  else:
    return a*(fact(a-1))

print(fact(5))

#Permutations of n objects taken r at a time
def perm(n,r):
  b = fact(n)
  c = fact(n-r)
  d = b/c
  return d 

print(perm(8,4))


#Combinations of i objects taken j at a time
def comb(i,j):
  e = fact(i)
  f = fact(i-j)
  g = fact(j)
  h = e/(f*g)
  return h 

print(comb(8,4))
```

**Write a function that converts a decimal number to binary number**

```
a = int(input("Enter a positive integer value - "))
binary = []
while a > 0:
  if a % 2 == 0:
    a = a / 2
    binary.append(0)
  else:
    a = (a-1)/2
    binary.append(1)

binary.reverse()
for e in binary:
  print(e, end='')
```

**Write a function sumPdivisors() that finds the sum of proper divisors of a number. Proper divisors of a number are those numbers by which the number is divisible, except the number itself.  For example proper divisors of 36 are 1, 2, 3, 4, 6, 9, 18**

```
a = int(input("Please enter any positive number - "))
def sumPdivisors(a):
  lst = []
  b = a // 2
  for ele in range(1,(b+1)):
    if a % ele == 0:
      lst.append(ele)

  return lst

print('Proper divisors of {} are - {}'.format(a,sumPdivisors(a)))
```

**A number is called perfect if the sum of proper divisors of that number is equal to the number. For example 28 is perfect number, since 1+2+4+7+14=28. Write a program to print all the perfect numbers in a given range**

```
def sumPdivisors(a):
    ''' SUM OF PROPER DIVISORS '''
    lst = []
    b = a // 2
    for ele in range(1, (b + 1)):
        if a % ele == 0:
            lst.append(ele)
    s = sum(lst)
    return s


n = int(input('enter any max range value - '))


def perfNum(n):
    lst2 = []
    for ele in range(1, (n + 1)):
        if ele == sumPdivisors(ele):
            lst2.append(ele)
    print("The perfect numbers from 1 to {} are \n{}".format(n, lst2))


perfNum(n)

```

**Two different numbers are called amicable numbers if the sum of the proper divisors of each is equal to the other number. For example 220 and 284 are amicable numbers. 
Sum of proper divisors of 220 = 1+2+4+5+10+11+20+22+44+55+110 = 284
Sum of proper divisors of 284 = 1+2+4+71+142 = 220
Write a function to print pairs of amicable numbers in a range**

```
val = int(input('Enter a max range value - '))


def sumPdivisors(a):
    ''' SUM OF PROPER DIVISORS '''
    lst = []
    b = a // 2
    for ele in range(1, (b + 1)):
        if a % ele == 0:
            lst.append(ele)
    s = sum(lst)
    return s



def findPair(val):
    lst2 = []
    for ele in range(1, (val + 1)):
        c = sumPdivisors(ele)
        d = sumPdivisors(c)
        if (d == ele) and (c != d):
            lst2.append(ele)
    print('List of amicable numbers within {} are - {}'.format(val,lst2))
    


print(findPair(val))
```

**Write a program which can map() and filter() to make a list whose elements are cube of even number in a given list**

```
a = int(input('enter max value of range - '))
lst = []
for ele in range(1, (a+1)):
  lst.append(ele)

print(lst)

even = list(filter(lambda x:(x % 2 == 0),lst))
cubes = list(map(lambda n: n**3, even))

print(cubes)
```
